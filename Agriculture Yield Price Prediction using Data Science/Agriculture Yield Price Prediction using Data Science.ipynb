# Label Encoder 

```python
data=pd.DataFrame()
from sklearn.preprocessing import LabelEncoder
col=["Soil_Moisture","Temperature","Crop_Name","Sunlight","Humidity","Region-1","Region-2","Region-3","Region-4"]
n1=LabelEncoder()
for i in col:
    data[i]=n1.fit_transform(data1[i])
```

# Model

```python
x=data[["Soil_Moisture","Temperature","Crop_Name","Sunlight","Humidity"]]
y=data1[["Profitable"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,shuffle=True,random_state=0,test_size=0.3)
```

```python
def model_Train(model,x,y):
    model = model
    model.fit(x,y)
    yp=model.predict(x)
    from sklearn.metrics import confusion_matrix, accuracy_score,recall_score,f1_score,precision_score
    cm = confusion_matrix(y, yp)
    
    accuracy = accuracy_score(y,yp)*100
    print('Training Accuracy of our model is equal ' + str(round(accuracy, 2)) + ' %.')
```

```python
def model_Test(model,x,y_test):
    model = model
    model.fit(x,y_test)
    y_pred = model.predict(x)

    from sklearn.metrics import confusion_matrix, accuracy_score,recall_score,f1_score,precision_score
    cm = confusion_matrix(y_test, y_pred)

    accuracy = accuracy_score(y_test, y_pred)*100
    print('Testing Accuracy of our model is equal ' + str(round(accuracy, 2)) + ' %.')
```

#  Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(max_iter=2000)
model.fit(x_train,y_train)
model_Train(model,x_train,y_train)
model_Test(model,x_test,y_test)
```

# SVC

```python
from sklearn.svm import SVC
model = SVC(kernel='linear') 
model_Train(model,x_train,y_train)
model_Test(model,x_test,y_test)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
model= GaussianNB()
model_Train(model,x_train,y_train)
model_Test(model,x_test,y_test)
```

# Lasso Regression

```python
from sklearn.linear_model import Lasso
model = Lasso(max_iter=10000)
model.fit(x_train, y_train)
y_pred = pd.DataFrame(model.predict(x_test))
print("Accuracy: ",model.score(x_test,y_pred)*100)
```

# Ridge Regression 

```python
from sklearn.linear_model import Ridge
model = Ridge(max_iter=1000)
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print("Accuracy: ",model.score(x_train,y_train)*100)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier,ExtraTreesClassifier
model=RandomForestClassifier()
model_Train(model,x_train,y_train)
model_Test(model,x_test,y_test)
```
